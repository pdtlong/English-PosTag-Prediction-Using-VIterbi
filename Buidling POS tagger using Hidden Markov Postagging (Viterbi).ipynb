# Read data

```python
import os
import fnmatch
import pandas as pd
files_tagged =[]
top = os.getcwd()
for path,dirs,files in os.walk(top):
    for file in files:
        if fnmatch.fnmatch(file,'*.POS'):
            fullname = os.path.join(path,file)
#             print(fullname)
            with open(fullname,"r") as f:
                files_tagged+=[f.read().splitlines()[17:]]
    print("Số file đã ở các thư mục con: ",len(files))
```

```python
#in thư 1 file đã được đọc
print(files_tagged[0])
```

```python
from itertools import groupby

#group các file đã đọc được vào 1 list
sents_tagged = [j for i in files_tagged for j in i]
```

```python
#Số dòng hiện tại
print("Số dòng đã đọc được: ",len(sents_tagged))
```

# Clean data

```python
#Xóa dấu xuống dòng và ký tự rỗng
sents_tagged = [x for x in sents_tagged if x not in "''"]

# xóa ký tự "[..]" và khoảng trống dư thừa
for i,sent in enumerate(sents_tagged):
    sents_tagged[i] = sent.replace("[ ", "").replace(" ]", "").replace("======================================", "").rstrip()
```

```python
print("Kết quả sau khi xóa ký tự [ , ]: \n",sents_tagged[0:30])
```

Ngắt câu

```python
split_sent_char="=================================="
                
sents_tagged_splited = [list(j) for i, j in groupby(sents_tagged, lambda x:x == split_sent_char) if not i] 
```

```python
sents_tagged_splited
```

```python
#Chuyển các câu trên thành list: sent [( word1,pos_tag1), (word2/pos_tag)..]
sents_tagged_clean=[]
for sent in sents_tagged_splited:
    temp1=[]
    for line in sent:
        temp1+=line.split(" ")
        temp2=[]
        for word in temp1:
            split_word=tuple(map(str, word.split('/'))) 
            if len(split_word) == 2:
                temp2.append(split_word)
    #Thêm câu vừa được xử lý vào copus
    sents_tagged_clean.append(temp2)
    temp2=[]
```

# Chuyển Data đã clean sang data train & test

- '##' là ký hiệu bắt đầu câu
- '&&' là ký hiệu kết thúc một câu

```python
mod_sent_tag=[]
for s in sents_tagged_clean:
    s.insert(0,('##','##'))
    s.append(('&&','&&'))
    mod_sent_tag.append(s)
```

Chia tập train và test

```python

split_num = int(len(mod_sent_tag)*0.8)
train_data = mod_sent_tag[0:split_num]
test_data = mod_sent_tag[split_num:]
```

Tạo từ điển cho các TAG
ex:- 'TAG':{word1: count(word1,'TAG')}

```python
train_word_tag = {}

for sent in train_data:
    for w,t in sent:
        w=w.lower()
        try:
            try:
                train_word_tag[t][w]+=1
            except:
                train_word_tag[t][w]=1
        except:
            train_word_tag[t]={w:1}
```

Tính xác suất của các từ nằm trong mỗi POS_TAG

```python
train_emission_prob={}
for k in train_word_tag.keys():
    train_emission_prob[k]={}
    count = sum(train_word_tag[k].values())
    for k2 in train_word_tag[k].keys():
        train_emission_prob[k][k2]=train_word_tag[k][k2]/count
```

```python
train_emission_prob
```

```python
#Ước lượng giá trị bigram giữa các từ trong câu
import nltk
bigram_tag_data = {}
for sent in train_data:
    #Lấy giá trị bigram bằng NLTK cho các từ trong câu
    bi_sent=list(nltk.bigrams(sent))
    for b1,b2 in bi_sent:
        try:
            try:
                bigram_tag_data[b1[1]][b2[1]]+=1
            except:
                bigram_tag_data[b1[1]][b2[1]]=1
        except:
            bigram_tag_data[b1[1]]={b2[1]:1}
```

```python
#Tính xác suất chuyển tiếp giữa các POS_TAG
bigram_tag_prob={}
for k in bigram_tag_data.keys():
    bigram_tag_prob[k]={}
    count=sum(bigram_tag_data[k].values())
    for k2 in bigram_tag_data[k].keys():
        bigram_tag_prob[k][k2]=bigram_tag_data[k][k2]/count
```

```python
bigram_tag_prob
```

```python
# tags_of_tokens dùng để ổng hợp các POS_TAG cho mỗi từ
# ví dụ: 'county': ['NNP', 'NN', 'NNPS']
tags_of_tokens = {}

for s in train_data:
    for w,t in s:
        w=w.lower()
        try:
            #Tổng hợp các pos_tag của mỗi từ trong training data
            if t not in tags_of_tokens[w]:
                tags_of_tokens[w].append(t)
        except:
            l = []
            l.append(t)
            tags_of_tokens[w] = l
            
for s in test_data:
    for (w,t) in s:
        w=w.lower()
        try:
            #Các POS_TAG có trong test data nhưng chưa tồn tại trong dictionary của trainning data
            #sẽ đc thêm vào nhầm tránh trường hợp 1 từ không có pos tag
            if t not in tags_of_tokens[w]:
                tags_of_tokens[w].append(t)
        except:
            l = []
            l.append(t)
            tags_of_tokens[w] = l
```

```python
tags_of_tokens
```

Chia label cho test set thành:
- test_sents chứa các câu cần dự đoán
- test_tags chứa các tag của câu đó

```python
test_sents=[]
test_tags=[]

for s in test_data:
    temp_word=[]
    temp_tag=[]
    for (w,t) in s:
        temp_word.append(w.lower())
        temp_tag.append(t)
        
    test_sents.append(temp_word)
    test_tags.append(temp_tag)
```

# Xây dựng thuật toán Viterbi cho từng câu

```python
def Predict_tag(sent): 
    #Dùng để lưu giá trị tag tốt nhất ở mỗi từ
    store_tag = {}
    #ex: store_tag = {step_i.:{tag_1:[Tag của từ phía trước , giá trị của Tag]}}

    for i,word in enumerate(sent):

        #Đây là trường hợp dùng để gán xác suất thấp nhất cho các ký tự đặc biệt '##' đầu câu)
        if i == 1:
            store_tag[i] = {}

            tags = tags_of_tokens[word]
            for t in tags:
                try:
                    # Nếu từ trong tập test tồn tại trong train data
                    store_tag[i][t] = ['##',bigram_tag_prob['##'][t]*train_emission_prob[t][word]]
                except:
                    # '##' không nằm trong corpus của train data lúc này xác suất sẽ là 0.0001
                    store_tag[i][t] = ['##',0.0001]

        #Duyệt từ đầu tiên
        if i>1:
            store_tag[i] = {}

            #Load tag của word thứ i-1
            previous_tags = list(store_tag[i-1].keys())

            #Duyệt lần lượt các tag của từ thứ i
            current_tags  = tags_of_tokens[word]
            
            #tìm giá trị tốt nhất của tag i-1 so với các tag ở i và lưu vào store_tag
            for t in current_tags:                             
                temp = []
                for pt in previous_tags:                         
                    try:
                        temp.append(store_tag[i-1][pt][1]*bigram_tag_prob[pt][t]*train_emission_prob[t][word])
                    except:
                        temp.append(store_tag[i-1][pt][1]*0.0001)
                max_temp_index = temp.index(max(temp))

                #Postag tốt nhất tìm được
                best_pt = previous_tags[max_temp_index]
                store_tag[i][t]=[best_pt,max(temp)]

    #Quay lui để tìm đường có giá trị cao nhất hay các tag phù hợp cho các từ trong câu
    pred_tags = []
    total_steps_num = store_tag.keys()
    last_step_num = max(total_steps_num)
    
    for bs in range(len(total_steps_num)):
        step_num = last_step_num - bs
        
        if step_num == last_step_num:
            pred_tags.append('&&')
            pred_tags.append(store_tag[step_num]['&&'][0])
            
        if step_num<last_step_num and step_num>0:
            pred_tags.append(store_tag[step_num][pred_tags[len(pred_tags)-1]][0])

    return list(reversed(pred_tags))
```

Test hàm dự đoán

```python
Predict_tag('## he had his own face , dark and beautiful . &&'.split(" "))
```

Dự đoán postag trên toàn tập test

```python
#Thư viện dùng để tính thời gian chạy model
from tqdm.notebook import tqdm
# Khởi tạo biến kết quả
predicted_Pos_Tags = []

#Duyệt lần lượt từng câu i:
for sent in tqdm(test_sents):    
    predicted_Pos_Tags.append(Predict_tag(sent))
```

# Đánh giá mô hình

```python
#TÍnh toán độ chính xác của predicted so với test data
right = 0 
wrong = 0
for i in range(len(test_tags)):
    gt = test_tags[i]
    pred = predicted_Pos_Tags[i]
    for h in range(len(gt)):
        if gt[h] == pred[h]:
            right = right+1
    else:
        wrong = wrong +1 
print('Độ chính xác predicted đươc so với test data :: ',right/(right+wrong))
print('Giá trị loss so với test data: ',wrong/(right+wrong))
```
